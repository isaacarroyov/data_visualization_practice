# Visualizing Math: Line Integral (Vector Field + Function)

```python
import numpy as np
import matplotlib.pyplot as plt
```

El siguiente paso es definir los parámetros del campo vectorial y la curva

**Campo vectorial**
$$\vec{F}(x,y) = \sin(x)\hat{i} + \cos(y)\hat{j}$$

**Curva paramétrica**
$$\vec{r}(t) = t\hat{i} + \left( 5\left(t-2\right)^2 + 2\left(t-2\right)^3 \right)\, , 0\leq t \leq 2.8$$

# Plot 1 (Only vector field)

```python
#Domain of the vector field
X = np.linspace(  -0.25, 2.35, 25 )
Y = np.linspace(  -0.1, 5, 25 )


#Points in the XY plane
X,Y = np.meshgrid(X,Y)

#Value each component of the vector field => F(x,y) = u*i + v*j
U = np.sin(X)
V = np.cos(Y)

#magnitude of all the vectors
Magnitude = np.sqrt(U**2 + V**2)
```

```python
fig,ax = plt.subplots(figsize=(1920/300,1920/300),
                      facecolor = '#FFFCF3',
                      dpi=300
                     )
ax.axis("off")

ax.quiver( X, Y, #Position of each vector
           U, V, #value of each vector
           Magnitude, #magnitude of each vector, this will help the cmap
           width = 0.002,
           headwidth = 4, 
           headlength = 4,
           cmap = plt.cm.PuBuGn, #assigns a color to each vector depending its magnitude
          label = r'$\vec{F} = \sin(x)\hat{i} + \cos(y) \hat{j}$'
          )
ax.set_xlim(-0.28,2.4)
ax.set_ylim(-0.05,4.7)
```

# Plot 2 (Curve and direction)

```python
#--------- PARAMETRIC CURVE --------
#parameter t
t = np.linspace( -0.25, 2.35, 100 )

#parametric equation of the curve
X_curve = t
Y_curve = 5*( (t - 2)**2 ) + 2*( (t - 2)**3 )
```

```python
#------------------ r'(t) OF THE CURVE ---------------------------------------
U_dcurve = 1
V_dcurve = 4 - 14*t + 6*(t**2)
```

```python
fig,ax = plt.subplots(figsize=(1920/300,1920/300),
                      facecolor = '#FFFCF3',
                      dpi=300
                     )
ax.axis("off")

ax.quiver(X_curve, Y_curve,
          U_dcurve, V_dcurve,
          color = '#872108',
          width = 0.0025,
          headwidth = 3,
          headlength = 4
         )
ax.plot( X_curve, Y_curve,
         lw = 1.25,
         color = '#F24318'
        )
ax.set_xlim(-0.28,2.4)
ax.set_ylim(-0.05,4.7)
```
